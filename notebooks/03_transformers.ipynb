# Sub-Word Tokenization Fine-tuning
Training all transformer models fully: BPE (RoBERTa), WordPiece (BERT), and SentencePiece (ALBERT).


```python
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath('../src'))
from models.roberta import RobertaModel
from models.bert import BertModel
from models.albert import AlbertModel
```

```python
train_df = pd.read_pickle('../data/processed/train.pkl')
val_df = pd.read_pickle('../data/processed/val.pkl')

```

```python
# Train RoBERTa
roberta_model = RobertaModel(model_dir='../models/roberta')
roberta_model.train(train_df, val_df, epochs=3, batch_size=16)
```

```python
# Train BERT
bert_model = BertModel(model_dir='../models/bert')
bert_model.train(train_df, val_df, epochs=3, batch_size=16)
```

```python
# Train ALBERT
albert_model = AlbertModel(model_dir='../models/albert')
albert_model.train(train_df, val_df, epochs=3, batch_size=16)
```
