# Baseline LR Model
Training the character n-gram TF-IDF Logistic Regression model.


```python
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath('../src'))
from models.lr_tfidf import LRTfidfModel
from evaluation import compute_classification_metrics
```

```python
train_df = pd.read_pickle('../data/processed/train.pkl')
val_df = pd.read_pickle('../data/processed/val.pkl')
```

```python
model = LRTfidfModel(model_dir='../models/lr')
model.train(train_df, val_df)
```

```python
test_df = pd.read_pickle('../data/processed/test.pkl')
preds = model.predict(test_df['text'])
metrics = compute_classification_metrics(test_df['label'], preds)
print('Baseline Test Metrics:', metrics)
```
