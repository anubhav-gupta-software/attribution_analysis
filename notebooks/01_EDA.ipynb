# Exploratory Data Analysis
Let's load the data from `data/processed/` and look at class distribution and rationale lengths.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('../src')
from data_utils import *
```

```python
train_df = pd.read_pickle('../data/processed/train.pkl')
val_df = pd.read_pickle('../data/processed/val.pkl')
test_df = pd.read_pickle('../data/processed/test.pkl')
print(f'Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}')
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=train_df, x='label')
plt.title('Training Label Distribution (0: Non-Toxic, 1: Toxic)')
plt.show()
```
