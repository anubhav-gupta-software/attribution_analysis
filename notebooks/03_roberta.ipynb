# RoBERTa Fine-tuning
Training the transformer model.


```python
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath('../src'))
from models.roberta import RobertaModel
```

```python
train_df = pd.read_pickle('../data/processed/train.pkl')
val_df = pd.read_pickle('../data/processed/val.pkl')

# We take a small sample to show it works quickly. Remove .head() for real training
train_sample = train_df.head(500)
val_sample = val_df.head(100)
```

```python
model = RobertaModel(model_dir='../models/roberta')
model.train(train_sample, val_sample, epochs=1, batch_size=16)
```

```python
# For evaluation, load best model
test_df = pd.read_pickle('../data/processed/test.pkl')
model.load()
preds, probs = model.predict(test_df.head(100)['text'].tolist())
from evaluation import compute_classification_metrics
print('RoBERTa Validation Sample Metrics:', compute_classification_metrics(test_df.head(100)['label'], preds))
```
