# RoBERTa Fine-tuning
Training the transformer model fully.


```python
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath('../src'))
from models.roberta import RobertaModel
```

```python
train_df = pd.read_pickle('../data/processed/train.pkl')
val_df = pd.read_pickle('../data/processed/val.pkl')

```

```python
model = RobertaModel(model_dir='../models/roberta')
model.train(train_df, val_df, epochs=3, batch_size=16)
```

```python
# For evaluation, load best model
test_df = pd.read_pickle('../data/processed/test.pkl')
model.load()
preds, probs = model.predict(test_df['text'].tolist(), batch_size=16)
from evaluation import compute_classification_metrics
print('RoBERTa Test Metrics:', compute_classification_metrics(test_df['label'], preds))
```
